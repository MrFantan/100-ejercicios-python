### **Ejercicio 8: Generación de números mediante bucles**

Escribe un programa en Python que permita imprimir en la consola los números del **1** al **20**.

- **Nota**: es necesario crear dos versiones, una con el bucle _**for**_ y la otra con el bucle _**while**_.

````python
# Version For
print('Inicio el bucle for')
n = []

for n in range(1,21):
    print(n)
    

print('Fin el bucle for')
print('----------------------')
print('Inicio el bucle While')
    
# Version While

n = 1
while n <= 20:
    print(n)
    n = n + 1

print('Fin el bucle While')
````

### **Ejercicio 9: Números impares incluidos en un intervalo**

Escribe un programa en Python que permita imprimir solo **los números impares** entre **10** y **20**.

- **Nota**: es necesario crear dos versiones, una con el bucle _**for**_ y la otra con el bucle _**while**_.

````python
# Version For
print('Inicio el bucle for')
n = []

for n in range(10,21):
    
    if n % 2 != 0:
        print(n, 'es un numero impar')

print('Fin el bucle for')
print('----------------------')
print('Inicio el bucle While')
    
# Version While

n = 10
while n < 20:
    if (n % 2) == 1:
        print(n,'es un número impar')
    n = n + 1
print('Fin el bucle While')
````

### **Ejercicio 17: Elementos en común entre dos listas**

Para probar, vamos a tomar las listas:

- _**L1 = [9, 8, 7, 14, 3, 2, "a", "p", "hola", "b"]**_
- _**L2 = ["b", 1, 9.2, 6, 3, 9, "p"]**_

````python
l1 = [9, 8, 7, 14, 3, 2, "a", "p", "hola", "b"]
l2 = ["b", 1, 9.2, 6, 3, 9, "p"]

l3 = [elementos for elementos in l1 if elementos in l2]

print(l3)
````

En este ejercicio, queremos realizar algunas operaciones con las instancias que crearemos a partir de la clase **Punto2D**. utilizando operadores estándar como **'+'**, **'-'**, **'\*'**, etc.

\

Para hacer esto, aquí está la lista de métodos especiales a crear para esta función:
  - **\_\_add\_\_(self, p)** que permitirá realizar la operación **p1 + p2**.
  - **\_\_sub\_\_(self, p)** que permitirá realizar la operación **p1 - p2**.
  - **\_\_mul\_\_(self, p)** que permitirá realizar la operación **p1 \* p2**.
  - **\_\_truediv\_\_(self, p)** que permitirá realizar la operación **p1 / p2**.
- **Recordatorio**: La sobrecarga de operadores consiste en crear métodos especiales en la clase que nos permitan utilizar los operadores estándar como **'+'**, **'-'**, **'\*'**, **'/'** yasí sucesivamente.

\

---

**Pruebas de verificación**:
```
p1 = Punto2D(3, 2)

p2 = Punto2D(1, 4)

p1 - p2:
>>> (2, -2)

p1 + p2:
>>> (4, 6)

p1 / p2:
>>> (3.0, 0.5)
```
